# Imports

Импортируем необходимые библиотеки

```python
import os
from pathlib import Path

from huggingface_hub import hf_hub_download, snapshot_download
```

Подгрузим .env

```python
HF_TOKEN = os.environ.get("HUGGINGFACE_HUB_TOKEN")
```

# Download IlyaGusev/saiga_mistral_7b_gguf

Скачаем модель

```python
FILENAME = "model-q4_K.gguf" 
HF_REPO = "IlyaGusev/saiga_mistral_7b_gguf"
OUT_DIR = Path(r"D:\Documents\ITMO\VKR\WideResearcher\models\saiga_gguf")
OUT_DIR.mkdir(parents=True, exist_ok=True)

print("HF repo:", HF_REPO)
print("Target dir:", OUT_DIR)
```

```python
print("Downloading repo snapshot (this may take a while)...")
local_path = hf_hub_download(
    repo_id=HF_REPO,
    filename=FILENAME,
    repo_type="model",
    local_dir=str(OUT_DIR),
    token=HF_TOKEN,
)
print("Downloaded:", local_path)
```
