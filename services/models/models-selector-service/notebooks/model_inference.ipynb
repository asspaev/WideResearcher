```python
url = "http://localhost:7000/v1/completions"

prompt = "Напиши короткий рассказ про осень"

data = {
    "model": "/models/misair-cotype-nano",
    "prompt": prompt,
    "max_tokens": 3000,    # больше токенов
    "temperature": 0.45,   # чуть ниже для связности
    "top_p": 0.9          # чтобы выбирать более вероятные варианты
}

response = requests.post(url, json=data)
response.raise_for_status()  # выбросит ошибку если что-то не так

result = response.json()

# vLLM обычно отдаёт текст в result['choices'][0]['text']
text = result.get("choices", [{}])[0].get("text", "")
print(text)
```
